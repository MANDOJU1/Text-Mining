```python
from konlpy.tag import Okt
okt = Okt()
```

```python
# 하나의 문자열로 처리

" ".join(df_news['message'])
```

```python
# 명사 추출

nouns = okt.nouns(" ".join(df_news['message']))
```

```python
from collections import Counter

count = Counter(nouns)
```

```python
count
```

```python
# count
# {'스파크': 3,
        #  '포럼': 4,
        #  '미래': 3,
        #  '시대': 7,}

# count.keys()
# count.values()
# for k, v in count.items():
#   print(k,v)


df_word = pd.DataFrame(
    {
        "word":count.keys(),
        "count":count.values()
    }
)
df_word
```

```python
df_word['word'].apply(lambda x:len(x) > 1)
```

```python
df_word[df_word['word'].apply(lambda x:len(x) > 1)].sort_values('count', ascending=False)[:50].plot.bar(x='word',
                                                                                                        y='count',
                                                                                                        figsize=(20,5),
                                                                                                        rot=40)
```

```python
from wordcloud import WordCloud

# 워드클라우드로 생성안했으면 하는 단어 추가
stop_word = ['및','과', '은','는']

font_path = r"/Library/Fonts/NanumBarunGothic.ttf"

wordcloud = WordCloud(font_path, stopwords=stop_word, background_color="ivory").generate(" ".join(df_word['word']))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
